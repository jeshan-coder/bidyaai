```python
import unsloth
import pandas as pd
import torch
from unsloth import FastLanguageModel
from peft import PeftModel # Used to load the PEFT adapter
import os
```

```python
max_seq_length = 2048
```

```python
model_id = "google/gemma-3n-e2b-it"
```

```python
print(f"Loading base model: {model_id}")
```

```python
model_id = "google/gemma-3n-e2b-it"
```

```python
path_to_checkpoint=r"/content/drive/MyDrive/gemma3n-finetune-checkpoints"
```

```python
base_model, tokenizer = unsloth.FastLanguageModel.from_pretrained(
    model_name=model_id,
    max_seq_length=max_seq_length,
    dtype=None, # Auto-detects dtype (bfloat16/float16) as during training
    load_in_4bit=True, # Must load in 4-bit as it was trained in 4-bit
)
print("Base model and tokenizer loaded.")
```

```python
from transformers import AutoProcessor
processor = AutoProcessor.from_pretrained(model_id)
```

```python
output_dir=path_to_checkpoint
```

```python
checkpoints = [d for d in os.listdir(output_dir) if d.startswith("checkpoint-")]
```

```python
if not checkpoints:
    print(f"Error: No checkpoints found in {output_dir}. Please ensure training completed successfully.")
    print("If you called trainer.save_model(), the weights might be directly in the output_dir.")
    # Fallback to output_dir if no checkpoints named 'checkpoint-XXX'
    adapter_path = output_dir
else:
    # Get the path to the latest checkpoint
    latest_checkpoint = max(checkpoints, key=lambda d: int(d.split('-')[1]))
    adapter_path = os.path.join(output_dir, latest_checkpoint)
```

```python
print(f"Loading LoRA adapter from: {adapter_path}")
# Load the LoRA adapter weights onto the base model
model = PeftModel.from_pretrained(base_model, adapter_path)
model.eval() # Set model to evaluation mode for inference
print("LoRA adapter loaded and prepared for inference.")
```

```python
# --- 3. Prepare Test Data ---
test_df = pd.read_csv("dataset_test.csv")
test_data = [{"query": q, "response": r} for q, r in zip(test_df["query"], test_df["response"])]
```

```python
def formatting_prompts_func_inference(query):
    return f"### Instruction:\n{query}\n\n### Response:\n"
```

```python
print("Test data loaded.")

# --- 4. Perform Manual Inference and Evaluation ---
print("\n--- Starting Manual Evaluation ---")
```

```python
generation_kwargs = {
    "max_new_tokens": 256, # Max length of the generated response
    "do_sample": True,    # Set to True for creative generation, False for deterministic (greedy)
    "temperature": 0.1,   # Controls randomness (lower = less random)
    "top_p": 0.5,         # Controls diversity (higher = more diverse)
    "pad_token_id": tokenizer.eos_token_id, # Or tokenizer.pad_token_id if different
    "eos_token_id": tokenizer.eos_token_id, # Model stops generating at this token
}
```

```python
output_excel_path = "evaluation_results.csv"
```

```python
evaluation_results = []
```

```python
for i, example in enumerate(test_data):
    query = example["query"]
    ground_truth_response = example["response"]

    # Prepare the input text for the model using the processor
    input_text = formatting_prompts_func_inference(query)
    # Use the processor to create inputs, as it handles multimodal model's expectations
    inputs = processor(text=input_text, return_tensors="pt").to(model.device) # <--- Use processor here

    # Generate response
    with torch.no_grad(): # Disable gradient calculation for inference
        generated_ids = model.generate(
            **inputs,
            **generation_kwargs,
        )

    # Decode the generated output
    # The generated_ids will include the input prompt. We need to extract only the generated part.
    generated_full_text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)

    # Extract only the model's predicted response by removing the input_text
    # Ensure to strip any leading/trailing whitespace
    predicted_response = generated_full_text.replace(input_text, "").strip()

    print(f"\n--- Example {i+1} ---")
    print(f"Query: {query}")
    print(f"Ground Truth: {ground_truth_response}")
    print(f"Generated: {predicted_response}")
    print("-" * 40)

    # Store results for Excel
    evaluation_results.append({
        "Query": query,
        "Ground Truth Response": ground_truth_response,
        "Predicted Response": predicted_response
    })

print("\n--- Manual Evaluation Complete ---")

# --- 5. Export Results to Excel ---
print(f"Exporting evaluation results to {output_excel_path}...")
results_df = pd.DataFrame(evaluation_results)
results_df.to_csv(output_excel_path, index=False)
print("Evaluation results saved to Excel successfully!")
```
